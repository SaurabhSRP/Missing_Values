#**4th technique - Count or Frequency Encoding**

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head() 
```

```python
columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
##checking all no. of labels
for features in train_set.columns[:]:
  print(features,':',len(train_set[features].unique()),'labels')
```

```python
train_set['Country'].value_counts()
```

```python
##lets convert the above data to dictionary
train_set['Country'].value_counts().to_dict()
##so basically we will encode valuecounts with the country names , this is frequency encoding, replace country name with its frequency
```

```python
##frequnecy encoding
country_map=train_set["Country"].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

###Advantages
####1) Not increasing feature space 

###Disadvantage
####1) not be able to handle when the frequency of a feature is same or provide the same weight if frequency are same 

#**5th Technique - Target Guided Oridnal Encoding**

###1) ordering the labels according to target
###2) Replace the labels by the joint probability of being 1 or 0 


```python
titanic=pd.read_csv("/content/drive/MyDrive/datasets/Titanic/train (1).csv",usecols=['Cabin','Survived'])
titanic.head()
```

```python
titanic['Cabin'].fillna('Missing',inplace=True)
```

```python
titanic.head(20)
```

```python
#all the cabin starts with some letter example C or G etc , lets extract those Cabin Alphabets 
titanic['Cabin']=titanic['Cabin'].astype(str).str[0]
```

```python
titanic.head()
```

```python
##another way to see
titanic['Cabin'].unique()
```

```python
titanic.groupby(['Cabin'])['Survived'].mean()
```

```python
titanic.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
titanic.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=titanic.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
##we will do mapping from now 
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
titanic['Cabin_ordinal_labels']=titanic['Cabin'].map(ordinal_labels2)
titanic.head()
```

#**6th technique - Mean Encoding**

```python
##basically in previous titanic dataset , the below code is mean encoding , its similar to target ordinal encoding but here we will encode the values with the mean
titanic.groupby(['Cabin'])['Survived'].mean()
```

#**7th Technique - Probability Ration Encoding**

```python
dat=pd.read_csv("/content/drive/MyDrive/datasets/Titanic/train (1).csv",usecols=["Cabin","Survived"])
dat.head()
```

```python
##Replacing 
dat['Cabin'].fillna('Missing',inplace=True)
```

```python
dat['Cabin'].unique()
```

```python
##we will just take FIrst letter of the those categories 
dat['Cabin']=dat['Cabin'].astype(str).str[0]
dat.head()
```

```python
 dat['Cabin'].unique()
```

```python
##Probability ration encoding
##find the percentage of survived
dat.groupby(['Cabin'])['Survived'].mean()
##we will consider as our probability
prob_dat=dat.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_dat=pd.DataFrame(prob_dat)
prob_dat
```

```python
##1-survived gives the percentage died , which we will use to find the survivals
prob_dat['Died']=1-prob_dat['Survived']
```

```python
prob_dat['Died']
```

```python
prob_dat.head()
```

```python
###Probabilty formula = Probability of survived / Probability of died
prob_dat['Probability_ratio']=prob_dat['Survived']/prob_dat['Died']
```

```python
prob_dat.head()
```

```python
##Convert this to a dictionary 
prob_dat["Probability_ratio"].to_dict()
##lets save it to a variable
probability_encoded=prob_dat["Probability_ratio"].to_dict()
```

```python
dat['Cabin_Encoded']=dat['Cabin'].map(probability_encoded)
```
